```python
# MultiEURLEX Dataset Exploration
from datasets import load_dataset
import numpy as np
from collections import Counter

# Load dataset
dataset = load_dataset('coastalcph/multi_eurlex', 'en', split='test', 
                       label_level='all_levels', trust_remote_code=True)

print(f"Total documents: {len(dataset)}")
print(f"Total labels: {len(dataset.features['labels'].feature.names)}")
```

```python
# Label distribution
all_labels = []
for doc in dataset:
    all_labels.extend(doc['labels'])

label_counts = Counter(all_labels)
labels_per_doc = [len(doc['labels']) for doc in dataset]

print(f"Total label assignments: {len(all_labels)}")
print(f"Labels per document: min={min(labels_per_doc)}, max={max(labels_per_doc)}, mean={np.mean(labels_per_doc):.1f}")
print(f"\nMost common 5 labels: {label_counts.most_common(5)}")
print(f"Least common 5 labels: {label_counts.most_common()[-5:]}")
```

```python
# Document length distribution
lengths = [len(doc['text'].split()) for doc in dataset]
print(f"Document length (words):")
print(f"  Min: {min(lengths)}, Max: {max(lengths)}")
print(f"  Mean: {np.mean(lengths):.0f}, Median: {np.median(lengths):.0f}")
```

```python
# Sample documents
for i in [0, 100, 500]:
    doc = dataset[i]
    print(f"\n{'='*60}\nDocument {i}: {doc['celex_id']}")
    print(f"Text: {doc['text'][:150]}...")
    print(f"Labels: {doc['labels'][:5]}{'...' if len(doc['labels']) > 5 else ''}")
```
